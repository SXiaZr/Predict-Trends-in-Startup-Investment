# Loading
This notebook is used to scrape data.

```python
# requests and parser packages
import requests
from bs4 import BeautifulSoup
import os
import re
import pickle
import pandas as pd
from InvestorDataFrame import InvestorDataFrame
DATA_PATH = './data/'
```

```python
# Load the list of investors, useless investors are marked as NONE
with open('./setting/investors', 'rb') as f:
    investors = pickle.load(f)
```

```python
# Start scraping
for i, investor in enumerate(investors):
    if type(investor)!=type(None):
        file = DATA_PATH + investor.replace(' ', '_') + '_pubdate.csv'
        if not os.path.exists(file):
            tk = InvestorDataFrame(investor)
            if tk._pipeline_flag == True:
                pass
            else:
                investors[i] = None
print('Done')
```

```python
# Save investors list to local and can be continued next time
with open('./setting/investors', 'wb') as f:
    pickle.dump(investors, f, protocol=pickle.HIGHEST_PROTOCOL)
```
